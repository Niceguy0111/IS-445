```python
demoPlot2 = dfUsa.plot.scatter(x='Long', y='Lat')
```

```python
countryFilter = ipywidgets.Textarea()
countryFilter
```

```python
stateFilter = ipywidgets.Dropdown(
    options = [ 'tx', 'hi', 'tn', 'ct', 'al', 'fl', 'ca', 'nc', 'ny', 'ky', 'mi',
       'ma', 'ks', 'sc', 'wa', 'co', 'nh', 'wi', 'me', 'ga', 'pa', 'il',
       'ar', 'mo', 'oh', 'in', 'az', 'mn', 'nv', 'ne', 'or', 'ia', 'va',
       'id', 'nm', 'nj', 'wv', 'ok', 'ri', 'vt', 'la', 'pr', 'ak', 'ms',
       'ut', 'md', 'mt', 'wy', 'sd', 'de', 'nd', 'dc' ]
)
```

```python
stateFilter
```

```python
import datetime as dt

dfUsState = dfUsa[ dfUsa.State == stateFilter.value ]
dfUsState[ 'Sighting date' ] = pd.to_datetime(dfUsState[ 'Sighting date' ])
dfUsState[ 'Sighting date' ].dt.year.plot.hist( title = stateFilter.value + ' Sightings per Year', xlabel = 'Sighting Year', bins = 50 )
```
