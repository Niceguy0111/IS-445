```python
import matplotlib.pyplot as plt

# Horizontal bar plot with gaps
fig, ax = plt.subplots()
ax.broken_barh([
    (2022+8/12, 4/12), 
    (2023+1/12, 4/12), 
    (2023+8/12, 4/12), 
    (2024+1/12, 1/12)
], (10, 9), facecolors='tab:orange')
ax.broken_barh([
    (2016+8/12, 4/12), 
    (2017+1/12, 4/12), 
    (2017+8/12, 4/12), 
    (2018+1/12, 4/12), 
    (2018+6/12, 1/12), 
    (2018+8/12, 4/12), 
    (2019+1/12, 4/12), 
    (2019+8/12, 4/12), 
    (2020+1/12, 4/12), 
    (2023+8/12, 4/12), 
    (2024+0/12, 1/12), 
    (2024+1/12, 1/12), 
    (2019+7/12, 2/12),          # Optum 1
    (2020+7/12, 2/12),          # Optum 2
    (2021+7/12, 2/12),          # Optum 3
    (2020+11/12, 1+1/12),       # Chipotle
    (2022+1/12, 5/12),          # USPS
    (2022+7/12, 2/12),          # Optum 4
    (2023+7/12, 2/12)           # Optum 5
], (20, 9), facecolors=(
    'tab:blue', 
    'tab:blue', 
    'tab:blue', 
    'tab:blue', 
    'tab:blue', 
    'tab:blue', 
    'tab:blue', 
    'tab:blue', 
    'tab:blue', 
    'tab:blue', 
    'tab:blue', 
    'tab:blue', 
    'tab:green', 
    'tab:green', 
    'tab:green',
    'tab:green', 
    'tab:green',
    'tab:green', 
    'tab:green',
))
ax.set_ylim(5, 35)
ax.set_xlim(2015, 2025)
ax.set_xlabel('Years in College')
ax.set_yticks([15, 25], labels=['Angie', 'David'])     # Modify y-axis tick labels
ax.grid(True)                                       # Make grid lines visible
ax.annotate('Covid', (2020+6/12, 25),
            xytext=(0.8, 0.9), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            fontsize=16,
            horizontalalignment='right', verticalalignment='top')

plt.show()
```
